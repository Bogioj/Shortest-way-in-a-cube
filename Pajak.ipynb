### Okreslmy zmienne

```python
#spider variables
x_pajaka = 1
y_pajaka = 2

#fly variables
x_muchy = 2
y_muchy = 3

#room dimensions
x = 3   #scania na lewo to 0, sciana na prawo to 3
y = 4   #sciana tylna to 4, sciana przednia to 0
z = 2 #zakladamy dla prostoty ze pajak idzie po scianie zawsze pionowo bo to najkrotsza droga zeby pokonac wymiar wysokosci
```

### Wazne zalozenie: po scianie w gore pajak zawsze idzie pionowo

### Tworzymy podzialke

```python
#dzielimy podloge i sufit na n elementow, w tym wypadku powiedzmy ze zrobimy podzialke 0.1
import numpy as np
podzialka = 0.01
x_array = np.arange(0,x+podzialka,podzialka)
y_array = np.arange(0,y+podzialka,podzialka)
print ("X: \n",x_array, "\n")
print ("Y: \n",y_array, "\n")
```

### Rozwazamy 4 scenariusze, pajak idzie na lewa, prawa, tylna lub przednia sciane. Iterujemy po kolei i liczymy odleglosci i wybieramy najmniejsza wartosc.

##### Tylna sciana

```python
#tylna sciana
import math

droga_calkowita_array1 = []  #tworzymy pusty array

for value in x_array:
    droga_podloga = math.sqrt((y-y_pajaka)**2+(value-x_pajaka)**2)
    droga_sciana = z
    droga_sufit = math.sqrt((y_muchy-y)**2+(x_muchy-value)**2)
    droga_calkowita = droga_podloga + droga_sciana + droga_sufit
    droga_calkowita_array1.append(droga_calkowita)

print(droga_calkowita_array1)
```

```python
#wybieramy minimalna wartosc
sciana_tylna = min(droga_calkowita_array1)
print("SCIANA TYLNA: \nMinimalna droga to:", round(sciana_tylna, 4), " czyli pierwiastek z: ", round(sciana_tylna**2,4))
```

##### Przednia sciana

```python
#przednia sciana
import math

droga_calkowita_array2 = []  #tworzymy pusty array

for value in x_array:
    droga_podloga = math.sqrt((0-y_pajaka)**2+(value-x_pajaka)**2)
    droga_sciana = z
    droga_sufit = math.sqrt((y_muchy-0)**2+(x_muchy-value)**2)
    droga_calkowita = droga_podloga + droga_sciana + droga_sufit
    droga_calkowita_array2.append(droga_calkowita)

print(droga_calkowita_array2)
```

```python
#wybieramy minimalna wartosc
sciana_przednia = min(droga_calkowita_array2)
print("SCIANA PRZEDNIA: \nMinimalna droga to:", round(sciana_przednia, 4), " czyli pierwiastek z: ", round(sciana_przednia**2,4))
```

##### Lewa sciana

```python
#lewa sciana
import math

droga_calkowita_array3 = []  #tworzymy pusty array

for value in y_array:
    droga_podloga = math.sqrt((value-y_pajaka)**2+(0-x_pajaka)**2)
    droga_sciana = z
    droga_sufit = math.sqrt((y_muchy-value)**2+(x_muchy-0)**2)
    droga_calkowita = droga_podloga + droga_sciana + droga_sufit
    droga_calkowita_array3.append(droga_calkowita)

print(droga_calkowita_array3)
```

```python
#wybieramy minimalna wartosc
sciana_lewa = min(droga_calkowita_array3)
print("SCIANA LEWA: \nMinimalna droga to:", round(sciana_lewa, 4), " czyli pierwiastek z: ", round(sciana_lewa**2,4))
```

##### Prawa sciana

```python
#prawa sciana
import math

droga_calkowita_array4 = []  #tworzymy pusty array

for value in y_array:
    droga_podloga = math.sqrt((value-y_pajaka)**2+(x-x_pajaka)**2)
    droga_sciana = z
    droga_sufit = math.sqrt((y_muchy-value)**2+(x_muchy-x)**2)
    droga_calkowita = droga_podloga + droga_sciana + droga_sufit
    droga_calkowita_array4.append(droga_calkowita)

print(droga_calkowita_array4)
```

```python
#wybieramy minimalna wartosc
sciana_prawa = min(droga_calkowita_array4)
print("SCIANA PRAWA: \nMinimalna droga to:", round(sciana_prawa, 4), " czyli pierwiastek z: ", round(sciana_prawa**2,4))
```

### Ktora sciana najlepsza i jaka najkrotsza droga? (przy zalozeniu ze pajak idzie w gore)

```python
dlugosci = {
    "sciana_tylna": sciana_tylna,
    "sciana_lewa": sciana_lewa,
    "sciana_prawa": sciana_prawa,
    "sciana_przednia": sciana_przednia
}
odp = min(dlugosci.values())
sciana = [var for var, value in dlugosci.items() if value == odp][0]
print("Najkrotsza droga to: ",sciana, round(odp, 4) , " czyli pierwiastek z: ", round(odp**2,4))
```

### Robimy podzialke dla podlogi, sciany i sufitu

```python
#dzielimy podloge, sufit i sciane na n elementow, w tym wypadku powiedzmy ze zrobimy podzialke 0.1
import numpy as np
podzialka = 0.01
x_array = np.arange(0,x+podzialka,podzialka)
y_array = np.arange(0,y+podzialka,podzialka)
z_array = np.arange(0,z+podzialka, podzialka)
print ("X: \n",x_array, "\n")
print ("Y: \n",y_array, "\n")
print ("Z: \n",z_array, "\n")
```

##### Iterujemy po krawedzi podlogi i dla kazdej iterowanej wartosci iterujemy po krawedzi sufita i liczymy odleglosc

```python
#tylna sciana
import math

droga_calkowita_array = []  #tworzymy pusty array

for value in x_array:
    droga_podloga = math.sqrt((y-y_pajaka)**2+(value-x_pajaka)**2)
    for value2 in z_array:

        droga_sciana = math.sqrt((value2-value)**2+(2-0)**2)
        droga_sufit = math.sqrt((y_muchy-value2)**2+(x_muchy-value)**2)
        droga_calkowita = droga_podloga + droga_sciana + droga_sufit
        droga_calkowita_array.append(droga_calkowita)


print(droga_calkowita_array)
```

```python
#wybieramy minimalna wartosc
droga_min = min(droga_calkowita_array)
print("Minimalna droga to:", round(droga_min, 4), " czyli pierwiastek z: ", round(droga_min**2,4))
```
